# Dataset 1
## Movies Dataset from Pirated Sites

## 1.数据集说明
下载数据集 [地址](https://www.kaggle.com/datasets/arsalanrehman/movies-dataset-from-piracy-website)   
数据集描述：该数据集是从一个盗版网站收集的，该网站每月约有 200 万访问者。 该数据包含好莱坞、宝莱坞、动漫等各行业的 20,000 多部电影。  

```python
df = pd.read_csv("/content/movies_dataset.csv", index_col=0)
print(len(df))
df.head(5)
```

该数据集的每一行表示一部电影的数据，共20548部电影。  
每一条电影的记录有14个属性，其中有标称属性:
+ appropriate_for
+ director
+ id
+ industry
+ language
+ posted_date
+ release_date
+ writer

有数值属性：
+ IMDb-rating
+ downloads
+ run_time
+ views

### 1.1 数据摘要和可视化

#### 1.1.1数据摘要
标称属性，给出每个可能取值的频数  
数值属性，给出5数概括及缺失值的个数

```python
nominal_attr = ["appropriate_for","director","id","industry","language","posted_date","release_date","writer"]
for col in nominal_attr:
    value = df[col].value_counts(dropna=False)
    print(col,":")
    print(value)
    value[:20].plot(kind="bar",figsize=(10,4))
    plt.show()
```

```python
numeric_attr = ["IMDb-rating","downloads","run_time","views"]
col = "IMDb-rating"
print(col,":")
value = df[col]
print(value)
print('Min:', value.quantile(0))
print('Q1:', value.quantile(0.25))
print('Q2:', value.quantile(0.5))
print('Q3:', value.quantile(0.75))
print('Max:', value.quantile(1))
print('缺失值个数:',value.isnull().sum())
value.plot(kind="box",figsize=(5,8))
plt.show()
print("\n\n")

col = "downloads"
print(col,":")
value = df[col]
print(value)
value = value.str.replace(',', '').astype(float)
print('Min:', value.quantile(0))
print('Q1:', value.quantile(0.25))
print('Q2:', value.quantile(0.5))
print('Q3:', value.quantile(0.75))
print('Max:', value.quantile(1))
print('缺失值个数:',value.isnull().sum())
value.plot(kind="box",figsize=(5,8))
plt.show()
print("\n\n")

col = "run_time"
print(col,":")
value = df[col]
print(value)
value = value.str.replace(' ', '')
value = value.str.replace('min', '')
value = value.str.replace('m', '')
def cal_time(time_str):
    if "h" in str(time_str):
        hours, minutes = str(time_str).split("h")
        if minutes!='':
            return int(hours) * 60 + int(minutes)
        else:
            return int(hours) * 60
    else:
        return time_str
value = value.apply(cal_time).astype(float)
print(value)
print('Min:', value.quantile(0))
print('Q1:', value.quantile(0.25))
print('Q2:', value.quantile(0.5))
print('Q3:', value.quantile(0.75))
print('Max:', value.quantile(1))
print('缺失值个数:',value.isnull().sum())
value.plot(kind="box",figsize=(5,8))
plt.show()
print("\n\n")

col = "views"
print(col,":")
value = df[col]
print(value)
value = value.str.replace(',', '').astype(float)
print('Min:', value.quantile(0))
print('Q1:', value.quantile(0.25))
print('Q2:', value.quantile(0.5))
print('Q3:', value.quantile(0.75))
print('Max:', value.quantile(1))
print('缺失值个数:',value.isnull().sum())
value.plot(kind="box",figsize=(5,8))
plt.show()
```

#### 1.1.2 数据缺失的处理
观察数据集中缺失数据，分析其缺失的原因。分别使用下列四种策略对缺失值进行处理:

将缺失部分剔除  
用最高频率值来填补缺失值  
通过属性的相关关系来填补缺失值  
通过数据对象之间的相似性来填补缺失值  
注意：在处理后完成，要对比新旧数据集的差异。

```python
for column in df.columns:
    print("属性名:", column,"缺失值个数:",df[column].isnull().sum())
```

数据集的各个列都存在缺失值，可能为数据收集过程中数据不完整

```python
# 将缺失部分剔除
df_1 = df.dropna()
print("原始数据集形状",df.shape)
print("删除缺失部分数据集形状",df_1.shape)
```

```python
# 用最高频率值来填补缺失值
df_2 = df.copy()
for column in df_2.columns:
    df_2[column].fillna(df[column].mode()[0], inplace=True)
print("原始数据集形状",df.shape)
print("新数据集形状",df_2.shape)
```

```python
# 通过数据对象之间的相似性来填补缺失值
from sklearn.impute import SimpleImputer
sim = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
sim.fit(df)
df_3 = pd.DataFrame(sim.transform(df), columns=df.columns)
print("原始数据集形状",df.shape)
print("新数据集形状",df_3.shape)
```
